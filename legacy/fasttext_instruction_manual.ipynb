```python
import numpy as np
from numpy import dot
from numpy.linalg import norm

def cos_sim(A, B):
    return dot(A, B)/(norm(A)*norm(B))

def euclidean_distance(A, B):
    return np.sqrt(np.sum((A - B) ** 2))
```
